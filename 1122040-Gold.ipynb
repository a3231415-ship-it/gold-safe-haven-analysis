```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 定義要抓取的金融標的 (對應變數設計)
# GC=F: 黃金期貨 (Target)
# DX-Y.NYB: 美元指數 (Input)
# ^TNX: 美國十年期公債殖利率 (Input)
# ^VIX: 恐慌指數 (Input)
tickers = ["GC=F", "DX-Y.NYB", "^TNX", "^VIX"]

# 設定時間範圍：涵蓋近年重要中東地緣政治事件的區間
start_date = "2022-01-01"
end_date = "2024-05-01"

print("正在從 Yahoo Finance 獲取資料...")
# 下載資料，我們主要取用 'Close' (收盤價)
data = yf.download(tickers, start=start_date, end=end_date)['Close']

# 重新命名欄位，讓後續分析的 DataFrame 更具可讀性
data.columns = ['DXY', 'Gold', 'US10Y', 'VIX']

# 檢視前五筆資料與資料缺失狀況
print("\n資料前五筆預覽：")
print(data.head())

print("\n資料缺失值統計：")
print(data.isnull().sum())
```
